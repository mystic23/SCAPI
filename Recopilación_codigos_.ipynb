#Gauss Seidel

```python
import numpy as np
import sympy as sp

A = np.array([[8,3,-3],[-2,-8,5],[3,5,-5]])
DP = np.diag(abs(A),0)
NDP = np.sum(abs(A),axis=1).T-DP

print('A = \n', A ,'\n\n','DP = \n', DP,'\n\n','NDP = \n', NDP ,'\n')


def cgjygs(A):
  DP = np.diag(np.abs(A),0)
  NDP = np.sum(np.abs(A),axis=1).T-DP
  if np.all(DP > NDP, axis=0):
    print("La matriz es estrictamente dominante en sentido diagonal")
    return True
  else:
     print("La matriz  No es estrictamente dominante en sentido diagonal")
     return False

print(cgjygs(A))
```

```python
ST = np.sum(A)
print('Suma Total\n',ST, '\n')

SC = np.sum(A,axis=0)
print('Suma Columnas\n', SC, '\n')

SF = np.sum(A,axis=1)
print('Suma Filas\n',SF,'\n')
```

```python
# GAUSS-JACOBI
x1,x2,x3 = sp.symbols('x1,x2,x3')

x=sp.Matrix([x1,x2,x3])
b=sp.Matrix([14, 5, -8])
A=sp.Matrix([[8,3,-3],[-2,-8,5],[3,5,10]])


print('SOLUCIÓN EXACTA\n')
print(sp.solve((A@x-b),x),'\n')
x_1=sp.solve((A@x-b)[0],x1)
x_2=sp.solve((A@x-b)[1],x2)
x_3=sp.solve((A@x-b)[2],x3)

x_1=sp.lambdify((x2, x3), x_1)
x_2=sp.lambdify((x1, x3), x_2)
x_3=sp.lambdify((x1, x2), x_3)

xk=sp.Matrix([[1],[-1],[1/2]])

tol=0.01

print('SOLUCIÓN APROXIMADA. Iteraciones\n')
print("%s, %.15s, %.15s, %.15s"%(' k',    'x1',    'x2',    'x3' ))

for k in range(1, 100):
  xk1=sp.Matrix([(x_1(xk[1],xk[2])),(x_2(xk[0],xk[2])),(x_3(xk[0],xk[1]))])
  Ea=abs(xk1-xk)
  #print(Ea, Ea[0])
  #print(Ea[0])
  # check if it is smaller than threshold
  dx = sp.sqrt(Ea.dot(Ea))
  print("%d, %.15f, %.15f, %.15f, %.15f, %.15f, %.15f"%(k, (xk1[0]), (xk1[1]), (xk1[2]),Ea[0],Ea[1],Ea[2]))
  if (Ea[0])<tol and (Ea[1])<tol and (Ea[2])<tol:
      #sw = True
      print('Converge!')
      break
  xk=xk1
```

#Interpolación Funcional

problema: Ddos un conjuto de puntos $n+1$ , en 2D encontrar el polinomio,

de grado $n$, que pasa por cada uno de ellos.

$P_{n}(X) = C_{n}X^n + C_{n-1}X^n-1 + ... +c_{1}X+c_{0}  ≠  0$

El polinomio de interpolación siempre es para estimar valores dentro de los intervalos en los que se construyo el polinomio

La interpolacion se utiliza cuando no tengo la expresión analitica de la función

y solo tengo los puntos



#Solución Matricial --> **Vandermonde**

Example Metodos numéricos con MatLab. Author John H.Mathews
image.png
image.png

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

def f(x):
  return x+2/x
#Dados $ n+1 $ puntos x,y
#x = input('... ')
#y = input('... ')

x=np.array([1, 2, 2.5])  #[1, 2, 2.5]
y=np.array([f(t) for t in x])

ejex=np.linspace(min(x),max(x))
coef_P=np.linalg.inv(np.vander(x))@y #Coeficientes del Polinomio de Interpolación
print(coef_P)
P=np.polyval(coef_P,ejex) # Pn(ejex)
plt.plot(x,y,'o',ejex,P)
Ea=abs(f(1.5)-np.polyval(coef_P,1.5))
print("Error aproximado: ", Ea)
```

#Solución de Sistemas de ecuaciones lineales mediante factorización de matrices, Doolittle, Crout y Cholesky

```python
#Sistema triangular
#U es sustitucion regresiva
import numpy as np, sympy as sp,matplotlib.pyplot as plt, scipy as sc

A = np.matrix([[1,-5,7,8],[-1,5,-7,9],[10,-2,11,8],[15,-55,70,11]])
b = np.array([2,-1,5,.5])
# print(np.eye(4,4), '\n') #explicacion de los intercambios
P, L, U = sc.linalg.lu(A)  #DOLITTLE-CROUNT
print('P= ', P, '\n\n',L,  '\n\n', U , '\n')
y= np.linalg.inv(L)@P@b.T

x_fm= np.linalg.inv(U)@y.T  #factorizacion de matrices
x_mc=np.linalg.inv(A)@b.T   #metodo clasico

print(x_fm,'\n\n', x_mc)
```
