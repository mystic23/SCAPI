#### Ejercicio

Dado un valor $n$, hallar la suma de los primeros $n$ términos de la sucesión de Fibonacci.

$$f_n=\cfrac{1}{\sqrt{5}}\left(\phi^n-\hat{\phi}^n\right),$$
donde
$\phi=\cfrac{1+\sqrt{5}}{2}$ y $\hat{\phi}=\cfrac{1+\sqrt{5}}{2}$.

```
n = 15 #int(input("Digite n: "))
L = np.arange(0, n, 1)
phi = (1 + np.sqrt(5))/2
cphi = (1 - np.sqrt(5))/2
F = (1/np.sqrt(5))*(phi**L-cphi**L)
print(F)
suma = np.sum(F)
print("La suma de los primeros %d términos de la sucesión de Fibonacci es %s" % (n, suma) )
```

$f_n=f_{\lfloor n/2\rfloor}+f_{\lceil n/2\rceil}+n,\,f_0=1,\,f_1=1\,n>0.$

$\lceil n/2\rceil$: Techo de $n/2$. Por ejemplo: Techo de 2.5 es 3.

$\lfloor n/2\rfloor$: Piso de $n/2$. Piso de 2.5 es 2.

```
from math import floor, ceil
def f(n):
  if n == 0 or n == 1:
    return 1
  else:
    return f(floor(n/2))+f(ceil(n/2))+n

for i in range(0, 50):
  print(f(i), end=', ')
```

Coupon-collector

```
m = 10 #int(input("Digite el número de canecas: "))
intentos = 100 #int(input("Digite el número de intentos: "))
L = [0]*intentos
for i in range(0, intentos):
  C = [False]*m
  cantidad_falsos = m
  lanzamientos = 0
  while cantidad_falsos > 0:
    r = int(np.random.uniform(low=0,high=m))
    if not C[r]: # Si la caneca está marcad en falso
      C[r] = True
      cantidad_falsos -= 1 #cantidad_falsos = cantidad_falsos - 1

    lanzamientos += 1

  L[i] = lanzamientos

media = np.mean(L)
desv_est = np.sqrt(np.var(L))
plt.text(50, 25, "media = {}".format(media))
plt.text(50, 15, "$\sigma= {} $".format(desv_est))
plt.hist(L)
plt.show()
```

### Pandas Dataframes

> "*Two-dimensional, size-mutable, potentially heterogeneous tabular data. Also contains labeled axes (rows and columns). Arithmetic operations align on both row and column labels. Can be thought of as a dict-like container for Series objects. The primary pandas data structure.*" - (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html)

```
import numpy as np
import pandas as pd
from pandas import ExcelWriter
A=[x for x in range(10)]
B=[]
C=[]
for x in A:
    B=B+['Nombre_'+str(A[x])]
    C=C+['Apellido_'+str(A[x])]

df1 = pd.DataFrame({'Id': A,
                   'Nombre': B,
                   'Apellido': C})
df1
```

Matplotlib has two forms of use
- Global
- Object Oriented

### Plotting forms

#### Global

In this form, we use the **Pyplot** module to customize the figure and handle most of the aspects for us. The Global approachment is easier for beginners but makes customization and complex graphs much more difficult.

```
plt.figure(figsize=(9,7), dpi=100)
sn.countplot(dataset['petal length (cm)'])
plt.xlabel("petal length (cm)")
plt.ylabel("Count")
plt.legend(["Countplot"])
plt.grid(True)
```

#### Object Oriented

In the OOP form, we can use each axis individually as an object. In this form, we have more control and the customization is easier for each axis and its attributes.

```
fig,ax = plt.subplots(1,3, figsize=(18,10))
sn.countplot(dataset['petal length (cm)'], ax=ax[0])
ax[0].set_xlabel('petal length')
sn.countplot(dataset['petal width (cm)'],ax=ax[1])
ax[1].set_xlabel('petal width')
sn.countplot(dataset['sepal width (cm)'], ax=ax[2])
ax[2].set_xlabel('sepal width');
```

# EDA

## What is EDA?

> *EDA or Exploratory Data Analysis, refers to the critical process of performing initial investigations on data so as to discover patterns,to spot anomalies,to test hypothesis and to check assumptions with the help of summary statistics and graphical representations.*

## Example

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
```

First, we get our dataset from the module.

```
df,label = sklearn.datasets.load_wine(return_X_y=True,as_frame=True)
df.insert(loc=13,column='target',value=label)
df['target'] = df['target'].astype('category')
df
```

The dataset consist of 13 columns numeric columns, being our the 'target' column our label.

**Quantitative variables**

For our quantitative variables, we will use at first one great tool provided by *Seaborn*, the pairplot. The pairplot will compare each numeric variable against each other, we will use it to observe relations between our variables.

```
fig=sns.pairplot(df)
```

```
fig
```

This plot will help us understand the relationship between numeric variables and their frequency using the histograms at the diagonal.

**Categorical variables**

We can get a lot of information about the relationship and tendencies using categorical variables.
In this case, we will use our label column as our hue to get structured information to obtain those mentiond tendencies



```
sns.catplot(data=df, x='target', y='alcohol', kind='bar')
```

Using the *Seaborn*'s catplot we can group data and get useful insights about the numeric variable using our categorical variables as a hue.

```
sns.catplot(data=df, x='target', y='ash', kind='point')
```

```
sns.catplot(data=df, x='target', y='ash', kind='violin')
```

```
sns.catplot(data=df, x='target', y='magnesium', kind='boxen'  )
```
