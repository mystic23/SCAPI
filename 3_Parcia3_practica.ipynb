```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
x = np.array([8,9,10,11,12])
y = np.array([17.453,21.460,25.752,30.301,35.084])
ejex = np.linspace(min(x), max(x), 100)

coef_P = np.linalg.inv(np.vander(x)) @ y  # Coeficientes del Polinomio de Interpolación
P = np.polyval(coef_P, ejex)  # Pn(ejex)

plt.plot(ejex, P, 'b--', label='Polinomio de Interpolación')
plt.plot(x, y, 'ko', label='Puntos obtenidos')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

# Imprimir la función de interpolación
func_interpolacion = np.poly1d(coef_P)

derivada = func_interpolacion.deriv()

print("La función de interpolación es:")
print(func_interpolacion)
print()

print("La derivada de la función de interpolación es:")
print(derivada)
print()

valores_derivada = derivada(x)
print("Los valores de la derivada en x son:")
print(valores_derivada)
```
