#Interpolaci贸n Funcional

El polinomio de Interpolaci贸n
Matricial(Vandermonde)

```python
import numpy as np, sympy as sp, matplotlib.pyplot as plt
def f(x):
  return x+2/x

x=np.array([1, 2, 2.5])
y=np.array([f(t) for t in x])
ejex=np.linspace(min(x),max(x))
coef_P=np.linalg.inv(np.vander(x))@y #Coeficientes del Polinomio de Interpolaci贸n
print(coef_P)
P=np.polyval(coef_P,ejex) # Pn(ejex)
plt.plot(x,y,'o',ejex,P)
Ea=abs(f(1.5)-np.polyval(coef_P,1.5))
```

El polinomio de Interpolaci贸n de Lagrange

```python
import numpy as np, sympy as sp, matplotlib.pyplot as plt
from scipy.interpolate import lagrange
def f(x):
  return x+2/x
x=np.array([1, 2, 2.5])
y=np.array([f(t) for t in x])
PL=lagrange(x,y)
# print(PL)
plt.title( PL)
ejex=np.linspace(min(x),max(x))
P=np.polyval(PL,ejex) # Pn(ejex)
plt.plot(x,y,'o',ejex,P)
Ea=abs(f(1.5)-np.polyval(PL,1.5))
plt.show()
```
