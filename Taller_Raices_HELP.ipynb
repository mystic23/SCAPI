```python
defdef regula_falsi(f, ak, bk, tol):
  print('\nMÉTODO DE egula Falsi\n')
  print('k\t ak\t bk\t ck\tEa\n')
  Ea=abs(bk-ak)
  k=0
  while Ea>=tol:
    ck=bk-(f(bk)*(bk-ak))/(f(bk)-f(ak))
    print(k,"\t",ak,"\t",bk,"\t\t",ck,"\t",Ea)
    x=np.linspace(ak,bk,1000)
    plt.grid()
    plt.plot(x,f(x),[ak,bk],[0,0],[ak,ak],[0,f(ak)],[bk,bk],[f(bk),0],[ck,ck],[0,f(ck)],[ak,bk],[f(ak),f(bk)])
    plt.text(ak,f(ak),'(ak,f(ak))')
    plt.text(bk,f(bk),'(bk,f(bk))')
    plt.text(ck,f(ck),'(ck,f(ck))')
    plt.legend()

    plt.title('Iteración {}'.format(k))
    plt.show()

    if f(ck)==0:
      print("El valor de la raíz es =", ck)
      return ck
    else:
      if f(ak)*f(ck)>0:
        ak=ck
      else:
        bk=ck
    Ea=abs(bk-ak)
    k+=1
  print("El valor de la raíz es =", ck)
  return ck
```

```python
def newton(f, df, xk, tol, maxiter):
  print('\nMÉTODO DE NEWTON\n')
  for k in range(maxiter):
    xk1 = xk - f(xk)/df(xk)
    Ea=abs(xk1 - xk)
    print(k,"\t\t",xk,"\t\t",Ea)
    if Ea<tol:
      print("El valor de la raiz es =", xk1)
      return xk1#[xk1,k]
    xk=xk1
  return [None]
```
