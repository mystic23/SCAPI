Ejercicio No 1. Leer y graficar cualquier funci√≥n

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
#np.linspace()    np.arange()
#3*x**2+.1*x-.5

x = sp.symbols('x')
f = sp.parse_expr(input("porfavor, digita la funcion f(x) = ")) ## leer lo que es una funci√≥n x*exp(-x**2/10)
print(sp.lambdify(x,f)(np.linspace(-1,3,5)))
func= sp.lambdify(x,f)
ejex = np.linspace(-1,3) #sino le digo cuentos me grafica 50
ejey = func(ejex)

plt.plot(ejex,ejey)
plt.xlabel('Eje x')
plt.ylabel('Eje y')

plt.title('Grafico de f(x)')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = sp.symbols('x')
f = sp.parse_expr(input("porfavor, digita la funcion f(x) = "))

df = f.diff(x)

ejex = np.linspace(-10, 10, 100)

ejey = sp.lambdify(x, f)(ejex)
df_ejex = sp.lambdify(x, df)(ejex)

x0 = float(input("Ingrese el valor de x0: "))
y0 = sp.lambdify(x, f)(x0)

#  pendiente de la recta tangente
m = float(sp.lambdify(x, df)(x0))


tangente_ejey = y0 + m * (ejex - x0)

# Graficar la funci√≥n y la recta tangente
plt.plot(ejex, ejey, label='f(x)')
plt.plot(ejex, tangente_ejey, label='Recta tangente')
plt.scatter(x0, y0, color='red', label='Punto (x0, f(x0))')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
#f(x) = x**2 + 2*x + 1
#x0 = -1
```

```python
from sympy.polys.polyroots import roots
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = sp.symbols('x')
f = sp.parse_expr(input("porfavor, digita la funcion f(x) = "))
Pn = sp.Poly(f)
coef_Pn = Pn.all_coeffs()
raices = (np.roots(coef_Pn)).real
ejex= np.linspace(-5,6) #(-5,6)
ejey=np.polyval(coef_Pn,ejex)
plt.plot(ejex,ejey,raices,np.zeros(len(raices)),'om')
#f(x)=(x+1)*(x-.5)**2*(x+7)**3
#f(x) = (x-1)*(x+3)**3*(x-5)**2
#f(x) = (x-1)*(x+3)**3*(x-5)**2  #cuando la multiplicidad es **2 es una recta tangente al eje x
plt.show()
```
