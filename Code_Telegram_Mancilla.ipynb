```python
import numpy as np, sympy as sp,matplotlib.pyplot as plt
#
x=sp.symbols('x')
f=sp.parse_expr(input('Por favor, digita la función f(x) = '))
Pn=sp.Poly(f)
coef_Pn=Pn.all_coeffs()
raices=(np.roots(coef_Pn)).real
print(raices)
ejex=np.linspace(-5, 6000)
ejey=np.polyval(coef_Pn,ejex)
plt.plot(ejex,ejey,raices,np.zeros(len(raices)),'om')
```

```python
# GAUSS-JACOBI
x1,x2,x3 = sp.symbols('x1,x2,x3')
x=sp.Matrix([x1,x2,x3])
b=sp.Matrix([14, 5, -8])
A=sp.Matrix([[8,3,-3],[-2,-8,5],[3,5,10]])
print('SOLUCIÓN EXACTA\n')
print(sp.solve((A@x-b),x),'\n')
x_1=sp.solve((A@x-b)[0],x1)
x_2=sp.solve((A@x-b)[1],x2)
x_3=sp.solve((A@x-b)[2],x3)
x_1=sp.lambdify((x2, x3), x_1)
x_2=sp.lambdify((x1, x3), x_2)
x_3=sp.lambdify((x1, x2), x_3)
xk=sp.Matrix([[1],[-1],[1/2]])
tol=0.01
print('SOLUCIÓN APROXIMADA. Iteraciones\n')
print("%s, %.15s, %.15s, %.15s"%(' k',    'x1',    'x2',    'x3' ))
for k in range(1, 100):
  xk1=sp.Matrix([(x_1(xk[1],xk[2])),(x_2(xk[0],xk[2])),(x_3(xk[0],xk[1]))])
  Ea=abs(xk1-xk)
  #print(Ea, Ea[0])
  #print(Ea[0])
  # check if it is smaller than threshold
  dx = sp.sqrt(Ea.dot(Ea))
  print("%d, %.15f, %.15f, %.15f, %.15f, %.15f, %.15f"%(k, (xk1[0]), (xk1[1]), (xk1[2]),Ea[0],Ea[1],Ea[2]))
  if (Ea[0])<tol and (Ea[1])<tol and (Ea[2])<tol:
      #sw = True
      print('Converge!')
      break
  xk=xk1
```
