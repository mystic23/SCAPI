```python
from math import factorial
import sympy as sp
def TaylorPoly(a: int, n: int) -> None:
    """
    En este c√≥digo, la funci√≥n TaylorPoly() calcula la serie
    de Taylor para la funci√≥n f(x) = variable f.
    Se utiliza la biblioteca sympy para trabajar con expresiones
    simb√≥licas. La funci√≥n toma dos argumentos: a (punto de expansi√≥n)
    y n (n√∫mero de t√©rminos en la serie de Taylor). La funci√≥n imprime
    el valor de la serie de Taylor.

    Args:
        a (int): Punto de expansi√≥n de la serie de Taylor.
        n (int): N√∫mero de t√©rminos en la serie de Taylor.

    Returns:
        None. Imprime el valor de la serie de Taylor.

    """
    x = sp.symbols('x')
    f  =  "AQUI VA LA SERIE QUE QUIERES CALCULAR"
    F = f
    T = f.subs(x, a)
    for k in range(1, n+1):
        dfk = sp.diff(f, x)
        T = T + dfk.subs(x, a)*((x - a)**(k))/factorial(k)
        f = dfk
    print("Valor de la serie:", sp.expand(T))

TaylorPoly(a=0, n=5) #Aqui el valor de a y n
```

```python
import sympy as sp
from math import factorial

def estimar_seno(a: int, error: float) -> float:
    """
    Estima el valor de la funci√≥n seno en el punto a, agregando t√©rminos a la serie de Taylor
    hasta que el error absoluto sea menor que el valor de error.

    Args:
        a (int): Punto de evaluaci√≥n.
        error (float): Valor m√°ximo del error absoluto permitido.

    Returns:
        float: Valor estimado de la funci√≥n seno en el punto a.

    """
    x = sp.symbols('x')
    f = sp.cos(x)
    F = f
    T = f.subs(x, a)
    k = 1
    while True:
        dfk = sp.diff(f, x)
        term = dfk.subs(x, a) * ((x - a) ** k) / factorial(k)
        T += term
        k += 1
        if abs(term.subs(x, 1)) < error:
            break
    return T.subs(x, 1).evalf()

# Estimar el valor de f(1) con un error absoluto menor que 10^-3
valor_estimado = estimar_seno(0, 1e-3)

print("El valor estimado de f(1) es:", valor_estimado)
```
